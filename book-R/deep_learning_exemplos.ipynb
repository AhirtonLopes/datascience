# Deep learning

## Sigmoid

```r
sigmoid = function(x) {
   1 / (1 + exp(-x))
}
x <- seq(-5, 5, 0.01)
plot(x, sigmoid(x), col='blue')
```

## Softmax

```r
softmax = function(x) {
   exp(x) / sum(exp(x))
}
x <- seq(-5, 5, 0.01)
plot(x, softmax(x), col='blue')
```
